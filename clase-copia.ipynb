<h4>2.3 Imputacion simple de valores numericos</h4>

```python
def calcular_promedio_int(variable):
    average = round(dfAutos[variable].astype('float').mean(axis=0))
    return average

def calcular_promedio_float(variable):
    average = round(dfAutos[variable].astype('float').mean(axis=0), 2)
    return average

def imputacion_numerica(variable, valor, dfAutos):
    if variable in dfAutos.columns:
        dfAutos[variable].replace(np.nan, valor, inplace=True)
        print("Imputacion realizada de forma correcta")
    else:
        print(f"la columna {variable} no existe en el set de datos")
```

```python
dfAutos.info()
```

```python
["normalized", "bore", "stroke", "horsepower", "peak"]
imputacion_numerica("normalized", calcular_promedio("normalized"), dfAutos)
imputacion_numerica("peak", calcular_promedio_int("peak"), dfAutos)
imputacion_numerica("bore", calcular_promedio_float("bore"), dfAutos)
imputacion_numerica("stroke", calcular_promedio_float("stroke"), dfAutos)
imputacion_numerica("horsepower", calcular_promedio_int("horsepower"), dfAutos)
imputacion_numerica("peak", calcular_promedio_int("peak"), dfAutos)
```

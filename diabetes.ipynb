```python
import pandas as pd
import numpy as np
```

```python
#columns
# 0. Number of times pregnant.
# 1. Plasma glucose concentration a 2 hours in an oral glucose tolerance test.
# 2. Diastolic blood pressure (mm Hg).
# 3. Triceps skinfold thickness (mm).
# 4. 2-Hour serum insulin (mu U/ml).
# 5. Body mass index (weight in kg/(height in m)^2).
# 6. Diabetes pedigree function.
# 7. Age (years).
# 8. Class variable (0 or 1).
cols_raw = [
    "Number of times pregnant",
    "Plasma glucose concentration",
    "Diastolic blood pressure",
    "Triceps skinfold thickness",
    "Two Hour serum insulin",
    "Body mass index",
    "Diabetes pedigree function",
    "Age",
    "Class variable"
]
cols = ['number_of_times_pregnant', 
        'plasma_glucose_concentration', 
        'diastolic_blood_pressure', 
        'triceps_skinfold_thickness', 
        'two_hour_serum_insulin', 'body_mass_index', 
        'diabetes_pedigree_function', 'age', 'class_variable']

# for i in cols_raw:
#     cols.append(i.lower().replace(" ","_"))
    
```

```python
df = pd.read_csv("diabetes.csv",names=cols)
df.describe()
```

```python
#en unas columnas 0 no deberia de existir entonces contamos los ceros en las columnas que no deberian de tener cero
num_missing = (df[['plasma_glucose_concentration', 
        'diastolic_blood_pressure', 
        'triceps_skinfold_thickness', 
        'two_hour_serum_insulin', 'body_mass_index']] == 0).sum()
print(num_missing)
```

```python
df[['plasma_glucose_concentration', 
        'diastolic_blood_pressure', 
        'triceps_skinfold_thickness', 
        'two_hour_serum_insulin', 'body_mass_index']] = df[['plasma_glucose_concentration', 
        'diastolic_blood_pressure', 
        'triceps_skinfold_thickness', 
        'two_hour_serum_insulin', 'body_mass_index']].replace(0, np.nan)
```

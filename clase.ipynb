<h4>1.3) Revisar los encabezados de los set de datos</h4>
<h5>
    Sanas practicas para nombrar las variables:
    <ol>
        <li>Los nombres de variables no deben de empezar con un numero</li>
        <li>Los nombres no pueden ser iguales a una palabra reservada (for, if, print)</li>
        <li>
            Los nombres no deben de contener espacios (engine size -> engineSize / engine-size)
        </li>
        <li>
            Los nombres de variables deben estar en el mismo idioma
        </li>
        <li>
            Los nombres de variables deben estar todos en Minuscula o Mayuscula
        </li>
    </ol>
</h5>

```python
print(f"Listado de variables dfAutos1: {dfAutos1.columns}")
print(f"Listado de variables dfAutos2: {dfAutos2.columns}")
```

<h4>1.4) Generar un unico set de datos</h4>
<h5>
    Puntos a considerar:
    <ol>
        <li>Las variables se llamen igual</li>
        <li>Contener la misma cantidad de variables y estar en el mismo orden</li>
